```python
import pandas as pd
import numpy as np
```

```python
hot100 = pd.read_csv("hot_100_extended.csv")
not_hot = pd.read_csv("not_so_hot_extended.csv")
```

```python
display(hot100.head())
display(hot100.shape)
```

```python
display(not_hot.head())
display(not_hot.shape)
```

```python
# Drop duplicates
# hot100.drop_duplicates(subset=["id"], keep="first", inplace=True, ignore_index=True)
# not_hot.drop_duplicates(subset=["id"], keep="first", inplace=True, ignore_index=True)
# Drop extra id column
hot100.drop(["id"], axis=1, inplace=True)
not_hot.drop(["id"], axis=1, inplace=True)
```

```python
display(hot100.shape)
display(not_hot.shape)
```

```python
#addint "hot" column to the dataframe
hot = ["Yes"]*100
hot100["hot"] = hot
hot100
```

```python
hot = ["No"]*3000
not_hot["hot"] = hot
not_hot
```

```python
# Concatinating 2 dataframes
all_songs = pd.concat([hot100, not_hot], axis=0)
display(all_songs.shape)
display(all_songs.head())
```

```python
from sklearn import datasets
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
all_songs.dtypes
```

```python
# Drop the columns that we won't use for clustering
X = all_songs.select_dtypes(include=np.number).drop(columns=["key", "mode", "duration_ms", "time_signature"])
X
```

```python
X.describe()
```

```python
#Scaling the data and saving the scaler:
import pickle

scaler = StandardScaler()
scaler.fit(X)
X_scaled = scaler.transform(X)
filename = "scaler.pickle" # Path with filename

#saving the scaler:
with open(filename, "wb") as file:
        pickle.dump(scaler,file)

X_scaled_df = pd.DataFrame(X_scaled, columns = X.columns)
display(X.head())
print()
display(X_scaled_df.head())
```

```python
#testing #k from 10 to 19:

K = range(2, 21)

inertia = []
silhouette = []

for k in K:
    print("Training a K-Means model with {} clusters!".format(k))
    kmeans = KMeans(n_clusters=k, random_state=1234, verbose=1)
    kmeans.fit(X_scaled_df)
    
    filename = "kmeans_" + str(k) + ".pickle"
    with open(filename, "wb") as file:
        pickle.dump(kmeans, file)

    inertia.append(kmeans.inertia_)
    silhouette.append(silhouette_score(X_scaled_df, kmeans.predict(X_scaled_df)))
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

fig, ax = plt.subplots(1,2,figsize=(16,8))
ax[0].plot(K, inertia, 'bx-')
ax[0].set_xlabel('k')
ax[0].set_ylabel('inertia')
ax[0].set_xticks(np.arange(min(K), max(K)+1, 1.0))
ax[0].set_title('Elbow Method showing the optimal k')
ax[1].plot(K, silhouette, 'bx-')
ax[1].set_xlabel('k')
ax[1].set_ylabel('silhouette score')
ax[1].set_xticks(np.arange(min(K), max(K)+1, 1.0))
ax[1].set_title('Silhouette Method showing the optimal k')
```

```python
# Based on Silhouette Method, K=3 looks better
```

```python
kmeans = KMeans(n_clusters=3, random_state=1234)
kmeans.fit(X_scaled_df)
```

```python
clusters = kmeans.predict(X_scaled_df)
pd.Series(clusters).value_counts().sort_index()
all_songs['clusters']=clusters
all_songs.tail()
```

```python
all_songs.to_csv('all_songs_clustered.csv',index=False)
```
